<font color='red' size=5><b>Сегментирование пользователей банка</b></font>

<div style="border:solid green 2px; padding: 20px">
Нам предстоит выполнить работу для одного из самых лучших банков страны «Метанпромбанк». Наша задача заключается в осуществлении <b>сегментирования пользователей банка по количеству потребляемых продуктов</b>. Сегментация покажет нам, как пользуются нашими услугами клиенты. В частности, мы сможем проанализировать то, как коррелируется данный признак и факт оттока пользователей, есть ли зависимость между количеством продуктов и заработной платой клиентов, полом и другими признаками. Таким образом, мы узнаем, насколько важным для банков является увеличение количества потребляемости продуктов одним клиентом, а также узнаем, какие категории клиентов чаще всего приобретают определенное количество продуктов банка.
<div>

## В ходе выполнения работы нам предстоит выполнить следующие задания:
    

- Проведение исследовательского анализа данных,
- Сегментирование пользователей на основе данных о количестве потребляемых продуктов,
- формулирование и проверка статистических гипотез:
    * проверим гипотезу различия дохода между теми клиентами, которые пользуются двумя продуктами банка, и теми, которые пользуются одним.
    * сформулируем и проверим статистическую гипотезу относительно представленных данных.

## В ходе выполнения данных заданий нам предстоит пройти следующие шаги:

<font color='orange' size=5><b>1) Шаг 1:</b></font>

**`Открыть файл с данными и изучить общую информацию`**


Для этого нам предстоит встроить сторонний файл с данными в проект. Мы сделаем это с помощью гугл-таблиц, в которые импортируем файл bank_dataset.csv, а далее с помощью ссылки на таблицу импортируем таблицу в наше рабочее пространство. После мы изучим общую информацию о таблице и ее внешнем виде и сделаем некоторые выводы о датасете.

<font color='orange' size=5><b>2) Шаг 2:</b></font>

**`Предобработка данных`**


Этот этап мы можем так же назвать подготовкой данных к анализу. На этом этапе мы будем:

- Заменить названия столбцов на удобные для нас, а также привести их к правильному стилю;

- Проверить пропуски. Если они есть, то необходимо откорректировать датасет. Выдвинем гипотезу о возможных причинах появления признаков. 

- Проверить тип данных. Если с ним есть проблемы, то привести данные к нужному типу.

- Проверить наличие дубликатов и удалить их в случае обнаружения. Сделать предположение о причинах их появления.

<font color='orange' size=5><b>3) Шаг 3:</b></font>

**`Исследовательский анализ данных`**


на данном этапе мы и проведем сегментацию. 

* мы сгруппируем наш датасет по признаку количества используемых продуктов. 

* посмотрим на среднее и медианное значение признаков для каждого полученного сегмента. Создадим портрет среднестатистического клиента для каждого клиента.

* построим матрицу корреляции для признаков и посмотрим на то, как взаимосвязаны наши признаки с количеством приобретенных продуктов.

* этот этап не обойтется без визуализации - построим графики распределения признаков (общих и для сегментов). Посмотрим на то, как распределены признаки и сделаем выводы.

<font color='orange' size=5><b>4) Шаг 4:</b></font>

**`Статистический анализ данных`**


На данном этапе мы будем проверять гипотезы. Сначала мы формулируем нулевую и альтернативные гипотезы, а затем проводим проверку нулевой гипотезы, заранее выбрав самостоятельно alpha. В ходе проверки мы должны будем проверить две гипотезы: одна из них дана нам в техническом задании, а вторую мы должны будем сами сформировать на основе полученных данных. Объясним, какой  критерий использовали для проверки гипотез и почему.

<font color='orange' size=5><b>5) Шаг 5:</b></font>

**`Подведение итогов и формирование рекомендаций для банка`**


На данном этапе мы обобщаем все результаты, полученные нами в ходе работы. На основе же этих данных мы дадим рекомендации банку о взаимодействии с клиентами, чтобы те не уходили или приносили больше прибыли.
